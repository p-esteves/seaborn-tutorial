Importando pacote

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

%reload_ext autoreload
%autoreload 2
```

Gráfico simples

```python
x_1 = np.linspace(0,5,10)
y_1 = x_1**2

plt.plot(x_1,y_1)

plt.title("Gráfico de dias ao quadrado")

plt.xlabel("Dias")
plt.ylabel("Dias ao quadrado")

#plt.show() pra mostrar fora e notebooks
```

Multiplos gráficos numa imagem

```python
plt.subplot(1,2,1)
plt.plot(x_1,y_1,'r')

plt.subplot(1,2,2)
plt.plot(x_1,y_1,'b')
```

Usando objetos figuras

```python
fig_1 = plt.figure(figsize=(5,4),dpi=100)
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9])

axes_1.set_xlabel('Dias')
axes_1.set_ylabel('Dias ao quadrado')
axes_1.set_title('Gráfico de dias ao quadrado')
axes_1.plot(x_1,y_1,label='x/x^2')
axes_1.plot(y_1,x_1,label='x^2/x')

axes_1.legend(loc=0)
#0 melhor automatico
#1 superior direito
#2 superior esquerdo
#3 inferior esquerdo
#4 inferior direito
#pode inputar uma tupla 

#grafico dentro do outro
axes_2 = fig_1.add_axes([0.45,0.45,0.4,0.3])
axes_2.set_xlabel('Dias')
axes_2.set_ylabel('Dias ao quadrado')
axes_2.set_title('Gráfico de dias ao quadrado')
axes_2.plot(x_1,y_1,'r')

#mensagem texto com origem 0,0
axes_2.text(0,40,'Message')
```

Subplots

```python
#matriz de graficos
fig_2, axes_2 = plt.subplots(figsize=(8,4),nrows=1,ncols=3)
#espaço entre plots
plt.tight_layout()

#acessar plots por índice
axes_2[1].set_title('Plot 2')
axes_2[1].set_xlabel('X')
axes_2[1].set_ylabel('X Quadrado')

axes_2[1].plot(x_1,y_1)
```
